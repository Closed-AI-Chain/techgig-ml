```python
data = data.drop(['DEVICETYPE','CHANNELTYPE','ID','USERAGENT','USERCITY','USERZIPCODE','USERPLATFORMUID','BIDREQUESTIP','PLATFORM_ID'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder

label_encoder = LabelEncoder()

# Assuming 'column' is the column you want to encode
data['PLATFORMTYPE'] = label_encoder.fit_transform(data['PLATFORMTYPE'])
```

```python
unique_keywords = []

# Iterate over each row in the 'KEYWORDS' column
for keywords in data['KEYWORDS']:
    # Split the keywords using the '|' delimiter and add them to the unique keywords list
    unique_keywords.extend(keywords.split('|'))

# Convert the list of unique keywords to a set to remove duplicates
unique_keywords = list(set(unique_keywords))

# Sort the list of unique keywords alphabetically
unique_keywords.sort()

# Print the list of unique keywords
# print(unique_keywords)
```

```python
data['KEYWORDS'] = data['KEYWORDS'].apply(lambda x: x.split('|'))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Create an instance of MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Fit the MultiLabelBinarizer to the column and transform the column
encoded_keywords = mlb.fit_transform(data['KEYWORDS'])

# Create a DataFrame with the encoded keywords
encoded_df = pd.DataFrame(encoded_keywords, columns=mlb.classes_)

# Concatenate the encoded DataFrame with the original data
data_encoded = pd.concat([data, encoded_df], axis=1)

# Print the encoded DataFrame
# print(data_encoded)
```

```python
data_encoded = data_encoded.drop(['KEYWORDS'],axis=1)
```

```python
data_encoded['TAXONOMY'] = data_encoded['TAXONOMY'].fillna('')
data_encoded['IS_HCP'] = data_encoded['IS_HCP'].fillna(0)
```

```python
data_encoded['URL'] = data_encoded['URL'].str.extract('www\.(.*?)\.com')
```

```python
data_encoded['URL'] = label_encoder.fit_transform(data['URL'])
```

```python
label_encoder.fit(data_encoded['TAXONOMY'])

# Transform the taxonomy column with label encoding
data_encoded['TAXONOMY'] = label_encoder.transform(data_encoded['TAXONOMY'])
```

Predicting HCP

```python
X = data_encoded.drop(['TAXONOMY','IS_HCP'], axis=1)
y = data_encoded['IS_HCP']
```

```python
X_train = X[:113937]
y_train = y[:113937]
X_test = X[113937:]
```

```python
#Render Forest
from sklearn.ensemble import RandomForestClassifier


# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=115)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# # Make predictions on the test set
y_pred_rf_hcp = rf_classifier.predict(X_test)
```

```python
feature_importances = rf_classifier.feature_importances_
```

```python
feature_importances
```

```python
# Get feature importances
importances = rf_classifier.feature_importances_

# Calculate the mean importance
mean_importance = importances.mean()

# Create a dataframe to store feature importances
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})

# Filter the dataframe based on importance threshold and drop columns
selected_features = feature_importance_df[feature_importance_df['Importance'] >= mean_importance]['Feature']
X_train_filtered = X_train[selected_features]
X_test_filtered = X_test[selected_features]

# Print the selected features
print("Selected Features:")
print(selected_features)
```
